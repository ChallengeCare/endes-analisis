```python
import lightgbm as lgbm
```

```python
cols_fit = ['nro_vivientes_jure', 'nro_vivientes_facto','region','area_residencia','fuente_agua',
                   'tipo_banho','tipo_piso','indice_riqueza',
                   'sexo','meses']
col_target = 'flag_anemia'
```

```python
parameters = {
    'application': 'multiclass',
    'objective': 'softmax',
    'metric': 'multi_logloss',
    'is_unbalance': 'true',
    'boosting': 'gbdt',
    'num_leaves': 31,
    'feature_fraction': 0.5,
    'bagging_fraction': 0.5,
    'bagging_freq': 20,
    'learning_rate': 0.05,
    'verbose': 0,
    'num_class': 4
}
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_anemia_complete.info()
```

```python
df_anemia_complete.flag_anemia.value_counts(dropna=False)
```

```python
df_anemia_complete.loc[:, df_anemia_complete.dtypes == 'object'] =\
    df_anemia_complete.select_dtypes(['object'])\
    .apply(lambda x: x.astype('category'))
```

```python
df_anemia_complete.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
feature= df_anemia_complete[cols_fit]
target= le.fit_transform(df_anemia_complete[col_target])
feature_train, feature_test, target_train, target_test= train_test_split(feature, target, test_size=0.20)
```

```python
train_data = lgbm.Dataset(feature_train, label=target_train)
test_data = lgbm.Dataset(feature_test, label=target_test)
```

```python
model = lgbm.train(parameters,
                       train_data,
                       valid_sets=test_data,
                       num_boost_round=5000,
                       early_stopping_rounds=100)
```

```python
ax = lgbm.plot_importance(model, max_num_features=40, figsize=(15,15))
```
